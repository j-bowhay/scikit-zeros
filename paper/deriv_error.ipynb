```python
import matplotlib.pyplot as plt
import numpy as np
import scipy

from skzeros import derivative

plt.style.use("style.mplstyle")
```

```python
zz = np.linspace(0, 5, 500)
res = scipy.interpolate.AAA(zz, np.sin(zz), clean_up=False)
res.support_points
```

```python
zzz = np.logspace(-17, 0)


def r_z(z):
    return derivative(res.support_points, res.support_values, res.weights, z)


plt.loglog(zzz, np.abs(res(zzz + 5) - np.sin(zzz + 5)) + 1e-16, label="$|r(z) - f(z)|$")
plt.loglog(zzz, np.abs(r_z(zzz + 5) - np.cos(zzz + 5)), label="$|r'(z) - f'(z)|$")
plt.ylabel("Absolute Error")
plt.xlabel("Distance From Support Point")
plt.legend()
plt.savefig("figures/log_deriv_zero_finding_accuracy.pdf")
```
